```python
#!python -m pip install pandas apyori
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv("fanituotteet.csv")
df.head(2)
```

```python
df.shape
```

```python
colnames = df.columns
colnames
```

```python
transactions = []
for i in range(0, df.shape[0]):
    transact =  [] 
    for j in range(1, df.shape[1]):
        if df.values[i, j]==1:
            transact.append(colnames[j])
    transactions.append(transact)
transactions[0:20]
#print(transactions[0:10])
```

```python
'''results = list(apriori(transactions = transactions, min_support = 0.1, min_confidnce = 0.1, min_lift = 1, min_length = 1, max_length = 5))
results
'''

# Tehtävä1:
# Kokeile muuttaa apriori-algometrin parametreja tässä notebookissa. Mitä tapahtuu tuloksille?
results = list(apriori(transactions = transactions, min_support = 0.02, min_confidnce = 0.9, min_lift = 2))#, min_length = 1, max_length = 5))
#results
```

```python
data = []
print("Rule Supp. Conf. Lift")
print(80*"-")

for a in results:
    rules = a.ordered_statistics
    support = a.support
    items = a.items
    
    # print(support)
    print(rules)
    for b in rules:
        lhs     = b.items_base
        rhs     = b.items_add
        conf    = b.confidence
        lift    = b.lift
        data.append([lhs, rhs, support, conf, lift])
        print("%s  ---> %s  %6.4f %6.4f %f" %(lhs, rhs, support, conf, lift))
        print("")

print("Number of rules is " + str(len(data)))
```

```python
columns = ["LHS", "RHS", "Support", "Confidence", "Lift"]
data1= pd.DataFrame(data, columns = columns)
data1
```

```python
data1.nlargest(8, columns = 'Confidence')
```

```python
data1.nlargest(8, columns = 'Support')
```

```python
data1.nlargest(8, columns = 'Lift')
```

# mlxtend apriori

```python
#!python -m pip install mlxtend
from mlxtend.frequent_patterns import apriori, association_rules

#poista ID-muuttuja

df = df.drop('ID', axis = 1)

# muuta tiedot 0/1-> boolean, mxlextend apriori antaa virheilmoituksen muuten

df_bool = df.astype(bool)

# Hae tyypillisimmät ostostapahtumat 

frequent_itemsets = apriori(df_bool, min_support = 0.02, use_colnames = True)

#Suodata ostotapahtumat sen mukaan, missä on eniten tuotteita, len()-hakee tapahtuman pituuden 
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)

frequent_itemsets.nlargest(10, 'length')
```

Määritä assosiaatiosäännöt

min_threshold : float (default: 0.8) Minimal threshold for the evaluation metric, via the metric parameter, to decide whether a candidate rule is of interest.

```python
import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore", RuntimeWarning)
```

min_threshold=0.8 vain säännön 'lift' >= 0,8 arvot otetaan huomioon selvitä mitä muita parametreja voi käyttää

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 0.8)
rules
```

```python
rules.nlargest(10, 'confidence')
```

```python
#!python -m pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
data2 = pd.DataFrame(te_ary, columns = te.columns_)
data2
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(data2, min_support = 0.02, use_colnames = True)
frequent_itemsets

print( "Length of Frequent itemsets is " + str(len(frequent_itemsets)) )

frequent_itemsets['Length'] = frequent_itemsets['itemsets'].apply(len)
frequent_itemsets.nlargest(10, 'Length')
```
